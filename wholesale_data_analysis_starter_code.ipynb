## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df["sub_total"] = df["qty"] * df["unit_price"]
df.head()
```

```python
# Create a column for shipping price.
df["total_weight"] = (df["unit_weight"] * df["qty"]).round(2)
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df["shipping_price"] = df.apply(lambda row: row["total_weight"] * 7.00 if row["total_weight"] > 50.00 else row["total_weight"] * 10.00, axis=1)
df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df["tax"] = ((df["sub_total"] + df["shipping_price"]) * .0925).round(2)
df["total_price"] = (df["sub_total"] + df["shipping_price"] + df["tax"]).round(2)
df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df["total_cost"] = ((df["unit_cost"] * df["qty"]) + df["shipping_price"]).round(2)
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df["profit"] = ((df["total_price"] - df["total_cost"]))
#df_sorted = df.sort_values(by="profit", ascending=False)
#df_sorted.head()
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
#sorted_top_5_clients_qty = df.sort_values(by='qty', ascending=False)
#top_5_clients = sorted_top_5_clients_qty['client_id'].head(5).unique()

for client_id in top_5_clients_df:
    total_amount_spent = df[df['client_id'] == client_id]['total_price'].sum()
    print(f"Client {client_id} spent a total of ${total_amount_spent:,.2f}.")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_5_client_data = [
    {
        'client_id': client_id,
        'total_units': df[df['client_id'] == client_id]['qty'].sum(),
        'total_shipping': df[df['client_id'] == client_id]['shipping_price'].sum(),
        'total_amount_spent': df[df['client_id'] == client_id]['total_price'].sum(),
        'total_profit': df[df['client_id'] == client_id]['profit'].sum()
    }
    for client_id in top_5_clients_df
]

top_5_clients_details = pd.DataFrame(top_5_client_data)
top_5_clients_details
```

```python
# Format the data and rename the columns to names suitable for presentation.
top_5_clients_details.rename(columns={"client_id" : "Client ID",
                                      "total_units" : "Total Units",
                                      "total_shipping": "Total Shipping Cost",
                                      "total_amount_spent" : "Total Spent",
                                      "total_profit" : "Total Profit"}, inplace=True)

# Define the money columns. 
## Total Shipping Cost, Total Spent, and Total Profit are money columns

# Define a function that converts a dollar amount to millions.
def convert_to_millions(currency_value):
    return round(currency_value / 1000000, 2)

# Apply the function to a column in your DataFrame

top_5_clients_details["Total Profit"] = top_5_clients_details["Total Profit"].apply(convert_to_millions)
top_5_clients_details["Total Spent"] = top_5_clients_details["Total Spent"].apply(convert_to_millions)
top_5_clients_details["Total Shipping Cost"] = top_5_clients_details["Total Shipping Cost"].apply(convert_to_millions)

# Rename the columns to reflect the change in the money format. 
top_5_clients_details.rename(columns={"Total Shipping Cost" : "Total Shipping Cost (Millions)",
                                      "Total Spent" : "Total Spent (Millions)",
                                      "Total Profit" : "Total Profit (Millions)"}, inplace=True)

top_5_clients_details
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sort_by_profit_df = top_5_clients_details.sort_values(by='Total Profit (Millions)', ascending=False)

sort_by_profit_df

```
